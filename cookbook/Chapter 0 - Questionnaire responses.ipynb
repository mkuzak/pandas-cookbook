```python
import pandas as pd
```

```python
df = pd.read_csv('../data/nlesc python workshop responses.csv', parse_dates=['Timestamp'])
```

```python
# Time of first response
t0 = df['Timestamp'][0]

# How many seconds after first response
speedSeconds = df['Timestamp'].apply(lambda t: (t - t0).total_seconds())

figure(figsize=(12,4))
subplot(1,2,1)
(speedSeconds/60).hist()
xlabel('Minutes')
ylabel('Responses')

subplot(1,2,2)
(speedSeconds/(60*60)).hist(bins=arange(0,30))
xlabel('Hours')
ylabel('Responses');
```

```python
byPyLevel = df.groupby('What is your level of Python')
byPyLevel.size().plot(kind='pie', autopct='%.2f %%')
axis('equal');
```

```python
byPdLevel = df.groupby('How much you know about Pandas')
byPdLevel.size().plot(kind='pie', autopct='%.2f %%')
axis('equal');
```

```python
byIpynbLevel = df.groupby('How experienced are you with iPython (and notebooks)')
byIpynbLevel.size().plot(kind='pie', autopct='%.2f %%')
axis('equal');
```

```python
pyLevel = [
    "What is Python",
    "I can read it, maybe...",
    "I'm ok with it, but there is always more to learn",
    "I dream in Python",
    "from Python import blackbelt",
]
df['PyLevel'] = df['What is your level of Python'].apply(lambda x: pyLevel.index(x))
df['PyLevel'].hist(bins=arange(0,6));
```

```python
pdLevel = [
    "They are black and white and cuddly",
    "It's a Python thing, right?",
    "I've been meaning to learn it for a while",
    "Everybody should be using it",
    "Is there another way to handle data?",
]
df['PdLevel'] = df['How much you know about Pandas'].apply(lambda x: pdLevel.index(x))
df['PdLevel'].hist(bins=arange(0,6));
```

```python
ipynbLevel = [
    "Wait, what?",
    "iAmNot",
    "Looks cool, but haven't really got into it",
    "Best invention since sliced bread",
    "It is the only thing I use",
]
df['ipynbLevel'] = df['How experienced are you with iPython (and notebooks)'].apply(lambda x: ipynbLevel.index(x))
df['ipynbLevel'].hist(bins=arange(0,6));
```
